```python
import pandas as pd
import numpy as np
import unicodecsv
titanic_df = pd.read_csv("titanic_data.csv")
with open('titanic_data.csv', 'rb') as f:
    reader = unicodecsv.DictReader(f)
    titanic_data = list(reader)
titanic_df
```

```python
"""Data Wrangling
   In titanic data there is some missing value in age of passengers
So, make calculation easier change the missing values by 0 """
titanic_df['Age'].fillna(0,inplace = True)
titanic_df
```

```python
titanic_df['Survived'].sum() #Using pandas dataframe
```

```python
"""Question 1. How many passengers survived on titanic?"""

group_by_survived = titanic_df.groupby('Survived').apply(len)
group_by_survived
```

```python
"""Question 2. Total Number of passengers on titanic on each class and out of them how many
                survive."""

group_by_class = titanic_df.groupby('Pclass').apply(len)
group_by_class
```

```python
group_by_class_survived = titanic_df.groupby(['Pclass','Survived']).apply(len)
group_by_class_survived 
```

```python
"""Q2 Answer. There are 216,184 and 491 passengers for first,second and third class respectively
               and out of them 136 passengers survive in first class, 87 survive in second class
               and 119 passengers survive in third class."""

"""Question 3. What factors made people more likely to survive?"""
mean_of_class_survived = titanic_df.groupby('Pclass')['Survived'].mean()
mean_of_class_survived
```

```python
group_by_sex = titanic_df.groupby('Sex').apply(len)
group_by_sex
```

```python
"""So first factor of survival is the passengers class, higher the class higher the percantage
 of survival. Let's find the another factor """
group_by_sex_survived = titanic_df.groupby(['Sex','Survived']).apply(len)
group_by_sex_survived 
```

```python
mean_of_sex_survived = titanic_df.groupby('Sex')['Survived'].mean()
mean_of_sex_survived 
```

```python
"""Survival count agewise but it also have a limitation , missing values in titanic data
is considered as 0."""
l = 0
def age_group(age):
    try:
        age = int(age)
        if age == 0:
            age = 'Missing'
        if age != 0 and age != 'Missing':
            lower = (age-age%10)
            age = str(lower+1) + ' to ' + str(lower+10)
        return age
    except:
        return 
    
# group_by_age_survived = titanic_df.groupby(['Age','Survived']).apply(len)
# group_by_age_survived
titanic_df['Age_group'] = titanic_df['Age'].apply(age_group)
group_by_age_survived 
```

```python
"""1st class passengers are more survived than 2nd and 3rd class passengers.Which show that
  rich person life is more valuable than poor person life but it is not fair."""
%pylab inline
import matplotlib as plt
import seaborn as sns
df = group_by_class_survived.reset_index()
p = sns.barplot(data = df, x = 'Pclass', y = 0, hue = 'Survived')
p.set_xticklabels(['First Class', 'Second Class', 'Third Class'])
p.set(xlabel = '', ylabel = 'Passenger Counts', title = 'Class Survival Counts')
```

```python
"""Female surviver are more than male surviver which show that female life is prefered than 
    man life."""
df = group_by_sex_survived.reset_index()
p = sns.barplot(data = df, x = 'Sex', y = 0, hue = 'Survived')
p.set_xticklabels(['Female', 'Male'])
p.set(xlabel = '', ylabel = 'Passenger Counts', title = 'Sex Survival Counts')
```

```python
"""Plot the survival count agewise but it also have a limitation ,which is some
   missing values in titanic data."""
df = group_by_age_survived.reset_index()
p = sns.barplot(data = df, x = 'Age_group', y = 0, hue = 'Survived')
p.set(xlabel = 'Age', ylabel = 'Passenger Counts', title = 'Agewise Survival Counts')
```

```python
p = sns.boxplot(data = titanic_df.dropna(subset = ['Age']).sort_values('Sex'),
                x = 'Sex', y = 'Age', hue = 'Survived')
p.set(title = 'Deck by Sex and Survival\nBox Plot', xlabel = '')
```
